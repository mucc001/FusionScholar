```python
import builtins
# A crawler which calls the semantic scholar API to get citation and reference info. 
timeouted = []
failed = []
json_errors = []
api_errors = []
counter = 0
for i in range(392, len(s2_doc)):
    counter += 1
    # Typically the API call limit is imposed after getting 165 files. 
    if counter % 165 ==0:
        print(i)
        time.sleep(210)
    docno = s2_doc[i]['docno']
    url = "https://api.semanticscholar.org/graph/v1/paper/{}?fields=externalIds,url,title,year,openAccessPdf,fieldsOfStudy,s2FieldsOfStudy,citations,references".format(docno)
    # store the response of URL
    try:
        response = urlopen(url)
    # Take care of 404 not found error
    except urllib.error.HTTPError as error:
        if error.code == 404:
            continue 
        # When the API call limit is imposed, record the failed index and sleep for 210s. 
        else:
            timeouted.append(i)
            time.sleep(210)
            continue
    except builtins.Exception as error:
        failed.append(i)
        api_errors.append((i,error))
        continue
    try:
        data_json = json.loads(response.read())
    except builtins.Exception as error:
        json_errors.append((i,error))
        continue
    paper_details_info.append(data_json)   
```

```python
import math
def calculate_nDCG(result, query_id, query_docno_gain):
    ideal_results = query_docno_gain[query_id]
    ideal_DCG = 0
    rank = 1
    for k, v in ideal_results.items():
        if rank == 1:
            ideal_DCG += v
        else:
            ideal_DCG += v / math.log2(rank)
        rank += 1
    if result[0] in ideal_results:
        nDCG = ideal_results[result[0]]
    else:
        nDCG = 0
    for i in range(1, len(result)):
        if result[i] in ideal_results:
            nDCG += ideal_results[result[i]] / math.log2(i + 1)
    nDCG /= ideal_DCG        
    return nDCG
```
